Let's extract the SDSS photometry at $z \sim 0.25$ ... 

```python
sdss_fields = ['PhotopropSDSS_u', 'PhotopropSDSS_g', 'PhotopropSDSS_r',  'PhotopropSDSS_i', 'PhotopropSDSS_z']
photo_sdss = ilsam.photo.load_snapshot_photoprop(basePath, 80, [[1, 1, 1]], fields = sdss_fields)

photo_sdss
```

```python
sub_fields = ['GalpropSfr', 'GalpropMstar', 'GalpropSatType',]
SAM_subhalos_80 = ilsam.groupcat.load_snapshot_subhalos(basePath, 80, [[1, 1, 1]], fields=sub_fields)
```

```python
plt.scatter(photo_sdss['PhotopropSDSS_u'], 
            SAM_subhalos_80['GalpropSfr'][SAM_subhalos_80['GalpropMstar'] * 1e9 >= 1e9],
            s=0.1)

plt.yscale('log')
plt.xscale('log')
plt.xlabel(r'$ {\rm SDSS \, u} \, [\mu {\rm Jy}]$', fontsize=14)
plt.ylabel(r'$\log_{10} {\rm SFR} \, [M_\odot {yr}^{-1}]$', fontsize=14)
```

```python
def uJy_to_ABmag(f_nu):
    return 22.5 - 2.5*np.log10(f_nu)
```

```python
plt.hist(uJy_to_ABmag(photo_sdss['PhotopropSDSS_r']), 100)
```

```python
plt.scatter(uJy_to_ABmag(photo_sdss['PhotopropSDSS_u']) - uJy_to_ABmag(photo_sdss['PhotopropSDSS_g']), 
            uJy_to_ABmag(photo_sdss['PhotopropSDSS_g']) - uJy_to_ABmag(photo_sdss['PhotopropSDSS_r']), 
            s=0.1, alpha=0.5)
```
